```python
diff_19 = score_campaigns_difference(ge19, ge17)
uns_19 = score_campaigns_uns(ge19, ge17)
mrp_19 = score_campaigns_mrp(ge19)
```

```python
incumbent = ge17.index[ ge17['winner'] == 'lab' ]
challenge = ge17.index[ ge17['winner'] != 'lab' ]
```

```python
ax = sns.distplot(diff_19.loc[incumbent,'difference'].dropna(),label='incumbent').set_title('GE17 - YouGovPoll')
ax = sns.distplot(diff_19.loc[challenge,'difference'].dropna(),label='challenge')
ax = sns.distplot(diff_19['difference'].dropna(),label='all')

plt.legend()
```

```python
ax = sns.distplot(mrp_19.loc[incumbent,'difference'].dropna(),label='incumbent').set_title('GE19 - YouGovPoll')
ax = sns.distplot(mrp_19.loc[challenge,'difference'].dropna(),label='challenge')
ax = sns.distplot(mrp_19['difference'].dropna(),label='all')

plt.legend()
```
